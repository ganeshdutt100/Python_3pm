```python
# Step 1: Import the built-in library for web browsers
import webbrowser
import time

# Step 2: Take user input for what they want to watch/search
print("--- Python Personal Assistant ---")
search = input("Which song or topic do you want to open on YouTube? ")

# Step 3: Create the YouTube search link
url = f"https://www.youtube.com/results?search_query={search}"

# Step 4: Open the browser automatically!
print("Opening your browser in 3 seconds...")
time.sleep(3) # Creates suspense!
webbrowser.open(url)

print("✅ Done! Enjoy your video.")
```
