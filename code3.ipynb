```python
import pyttsx3

engine = pyttsx3.init()

BLUE="\033[94m"
RESET="\033[0m"

print("Python Speaking Bot")
name = input("Enter a name from the audience ")
message = f"Hello {name}, welcome to this fun Python Demo"

print(f"\n{BLUE} Python is speaking , '{message}'{RESET}\n ")
engine.say(message)

engine.runAndWait()
```
