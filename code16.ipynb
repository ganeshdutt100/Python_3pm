```python
import qrcode
data =  'https://4achievers.com/'

img  = qrcode.make(data)
img.save('4achievers.jpg')
print("QR code generated and saved as 4achievers.jpg")
```

```python
import qrcode
data =  'https://www.youtube.com/'

img  = qrcode.make(data)
img.save('Youtube.jpg')
print("QR code generated and saved as youtube.jpg")
```

```python
import pyttsx3
engine = pyttsx3.init()
name  =  input("Enter your name: ")
message = f"Hello {name}, welcome to 4Achievers!"
print("....")
engine.say(message)
engine.runAndWait()
```

```python
import pyshorteners
url = 'https://www.youtube.com/watch?v=dQw4w9WgXcQ'
shortener = pyshorteners.Shortener()
short_url = shortener.tinyurl.short(url)
print(short_url)
```
