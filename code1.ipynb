```python
from tabulate import tabulate
items=  {"Apple":50, "Banana":100, "Orange":150}
cart = {}

for item in items:
    qty = int(input(f"{item} qty:"))
    if qty > 0:
        cart[item] = qty
# calculate total and prepare bill
total =  0
bill  =[]
for item,qty in cart.items():
    price = items[item]* qty
    total += price
    bill.append([item,qty, items[item],price])

print("\n Your Bill : ")
print(tabulate(bill, headers=["Item","Qty","Unit","Total"], tablefmt= "grid")) 
print(f"Total Amount : Rs {total}")   
```
